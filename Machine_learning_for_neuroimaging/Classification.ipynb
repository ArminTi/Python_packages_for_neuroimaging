```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import zscore
from sklearn.preprocessing import StandardScaler, MinMaxScaler, PolynomialFeatures
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
```

```python
data = pd.read_csv("PC_fC_cleaned_scale.csv")
data['Group_num'] = [1 if x == 'Control' else 2 for x in data['Group']]
Group = data['Group_num']
data.drop('Group_num', axis=1, inplace=True)
data
```

## Find features for predicting Group

Check for your train

```python
Feature = data.iloc[: , 4:]
X_train, X_test, y_train, y_test = train_test_split(Feature, Group, test_size=0.2, random_state=6)
```

```python
corr = X_train.corrwith(y_train)
corr.sort_values().round(2)
```

## SVM

Link: https://scikit-learn.org/stable/modules/svm.html

```python
from sklearn.svm import SVC
from sklearn.metrics import classification_report
```

```python
svm = SVC(C = 1, kernel = 'poly', degree = 3) #C = Regularization parameter. default  = 1, tradeoff = margin definition
# C > 1: hard margin, C < 1: low ,margin
svm.fit(X_train, y_train)
y_predict_train = svm.predict(X_train)
```

```python
y_predict_train
```

```python
acc = svm.score(X_train, y_train)
acc
```

```python
print(classification_report(y_train, y_predict_train))
```

```python
y_predict_test = svm.predict(X_test)
print(classification_report(y_test, y_predict_test))
```

## logistic Regression

Link: https://scikit-learn.org/stable/modules/linear_model.html#logistic-regression

```python
from sklearn.datasets import load_iris
from sklearn.linear_model import LogisticRegression
reg = LogisticRegression()
reg.fit(X_train, y_train)
```

```python
y_predict_train_LG = reg.predict(X_train)
print(classification_report(y_train, y_predict_train_LG))
```

```python
acc = reg.score(X_train, y_train)
acc
```

```python
y_predict_test_LG = reg.predict(X_test)
print(classification_report(y_test, y_predict_test_LG))
```
